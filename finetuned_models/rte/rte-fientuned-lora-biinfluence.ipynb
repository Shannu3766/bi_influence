```python
import os
import torch
import numpy as np
import evaluate
from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    DataCollatorWithPadding,
    TrainingArguments,
    Trainer
)
from peft import get_peft_model, LoraConfig, TaskType
os.environ["WANDB_DISABLED"] = "true"
```

```python
MODEL_ID = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
TASK = "rte"
LR = 2e-4
NUM_EPOCHS = 10
BATCH_SIZE = 8 
MAX_LEN = 256  
```

```python
dataset = load_dataset("glue", TASK)
tokenizer = AutoTokenizer.from_pretrained(MODEL_ID, use_fast=True)
```

```python
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right" 
```

```python
def preprocess_function(examples):
    inputs = [
        f"Premise: {p}\nHypothesis: {h}\nQuestion: Does the premise entail the hypothesis?"
        for p, h in zip(examples["sentence1"], examples["sentence2"])
    ]
    
    tokenized = tokenizer(
        inputs,
        truncation=True,
        max_length=MAX_LEN,
        padding="max_length" # Consistent padding
    )
    return tokenized

encoded_dataset = dataset.map(preprocess_function, batched=True)
```

```python
encoded_dataset = encoded_dataset.remove_columns(["sentence1", "sentence2", "idx"])
encoded_dataset = encoded_dataset.rename_column("label", "labels")
encoded_dataset.set_format("torch")
```

```python
model = AutoModelForSequenceClassification.from_pretrained(
    MODEL_ID,
    num_labels=2,
    device_map="auto",
    torch_dtype=torch.bfloat16, 
    trust_remote_code=True
)
```

```python
model.config.pad_token_id = tokenizer.pad_token_id
model.config.use_cache = False 
```

```python
peft_config = LoraConfig(
    task_type=TaskType.SEQ_CLS,
    r=64,                    
    lora_alpha=128,
    lora_dropout=0.05,
    bias="none",
    target_modules=[   
        "q_proj", "k_proj", "v_proj", "o_proj", 
        "gate_proj", "up_proj", "down_proj"
    ]
)
```

```python
model = get_peft_model(model, peft_config)
model.print_trainable_parameters()

metric = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

```python
from torch.utils.data import DataLoader
from transformers import DataCollatorWithPadding
from adaptive_lora.callbacks import AdaptiveLoRACallback

# -----------------------------
# 🧹 1. Clean up the dataset
# -----------------------------
if "label" in encoded_dataset["train"].column_names:
    encoded_dataset = encoded_dataset.rename_column("label", "labels")

# ✅ FIX 1: Removed 'token_type_ids' (Llama does not use this)
keep_cols = {"input_ids", "attention_mask", "labels"} 

for split in encoded_dataset.keys():
    drop_cols = [c for c in encoded_dataset[split].column_names if c not in keep_cols]
    if drop_cols:
        encoded_dataset[split] = encoded_dataset[split].remove_columns(drop_cols)

encoded_dataset.set_format(type="torch")

# -----------------------------
# 🧱 2. Data Collator & Loader
# -----------------------------
data_collator = DataCollatorWithPadding(tokenizer=tokenizer, pad_to_multiple_of=8)

train_dataset = encoded_dataset["train"]
eval_dataset = encoded_dataset["validation"]

val_dataloader = DataLoader(
    eval_dataset,
    batch_size=8,
    shuffle=False,
    collate_fn=data_collator
)


adaptive_callback = AdaptiveLoRACallback(
    val_dataloader=val_dataloader,
    total_rank=10000,  
    tau=0.7,
    verbose=True,
)

print("✅ Adaptive LoRA callback ready.")
```

```python


training_args = TrainingArguments(
    output_dir=f"{MODEL_ID}-RTE-v2",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=LR,
    per_device_train_batch_size=BATCH_SIZE,
    per_device_eval_batch_size=BATCH_SIZE,
    num_train_epochs=NUM_EPOCHS,
    weight_decay=0.01,
    logging_steps=20,
    load_best_model_at_end=True,
    metric_for_best_model="accuracy",
    fp16=False,             
    bf16=True,      
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=encoded_dataset["train"],
    eval_dataset=encoded_dataset["validation"],
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
    callbacks=[adaptive_callback],
)


print("🚀 Starting Training...")
trainer.train()
```

```python
print("📊 Final Evaluation:")
results = trainer.evaluate()
print(results)
```
