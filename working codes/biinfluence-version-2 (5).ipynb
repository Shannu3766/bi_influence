```python
import torch
from transformers import (
    AutoModelForCausalLM, 
    AutoTokenizer, 
    Trainer, 
    TrainingArguments
)
from datasets import load_dataset, Dataset
from peft import get_peft_model, LoraConfig, TaskType
from torch.utils.data import DataLoader
from adaptive_lora.callbacks import AdaptiveLoRACallback # Import our callback
```

```python
# --- 1. Setup Model and Tokenizer ---
```

```python
model_name = "mistralai/Mistral-7B-v0.1" # Example model
tokenizer = AutoTokenizer.from_pretrained(model_name)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(
    model_name, 
    torch_dtype=torch.bfloat16, # Use bfloat16 for efficiency
    device_map="auto"
)
```

```python
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    target_modules=["q_proj", "v_proj", "k_proj", "o_proj"],
    task_type=TaskType.CAUSAL_LM,
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
```

```python
# --- 4. Create the val_dataloader for the callback ---
# This dataloader provides the small batch for BI score computation
dataset = load_dataset("wikitext", "wikitext-2-raw-v1")

# Use only the first 50 examples (to keep it light)
train_texts = dataset["train"]["text"][:50]
val_texts = dataset["validation"]["text"][:20]

def tokenize_data(texts):
    enc = tokenizer(
        texts,
        truncation=True,
        padding="max_length",
        max_length=128,
        return_tensors="pt"
    )
    enc["labels"] = enc["input_ids"].clone()
    return Dataset.from_dict({k: v.tolist() for k, v in enc.items()})

train_dataset = tokenize_data(train_texts)
val_dataset = tokenize_data(val_texts)
```

```python

# ============================================================
# 4ï¸âƒ£ Prepare Validation Loader for Adaptive LoRA
# ============================================================
def collate_fn(batch):
    features = {key: [example[key] for example in batch] for key in batch[0].keys()}
    padded = tokenizer.pad(
        features,
        return_tensors="pt",
        padding="max_length",
        max_length=128
    )
    return padded

val_loader_for_callback = DataLoader(
    val_dataset,
    batch_size=4,
    collate_fn=collate_fn
)

adaptive_callback = AdaptiveLoRACallback(
    total_rank=512,
    tau=0.9,
    val_dataloader=val_loader_for_callback,
    # min_rank=4 has been REMOVED to match Algorithm 2
    verbose=True
)

# --- 6. Setup Trainer ---
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=4,
    logging_steps=10,
    report_to="none", # Disable wandb/etc. for this example
    save_strategy="no", # Don't save checkpoints
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset, 
    callbacks=[adaptive_callback], # Add our callback here
)


print("Starting training with Adaptive LoRA...")
trainer.train()

print("Training complete. Check logs/adaptive_lora_epoch_logs.csv for rank evolution.")
```
