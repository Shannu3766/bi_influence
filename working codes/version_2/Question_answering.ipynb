```python
import torch
from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForQuestionAnswering,
    TrainingArguments,
    Trainer,
    default_data_collator,
)
from peft import get_peft_model, LoraConfig, TaskType
from torch.utils.data import DataLoader
from adaptive_lora.callbacks import AdaptiveLoRACallback
import evaluate
import numpy as np
```

```python
# 1ï¸âƒ£ Load Model and Tokenizer
# ============================================================
model_name = "deepset/roberta-base-squad2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
```

```python
# ============================================================
# 2ï¸âƒ£ Apply LoRA
# ============================================================
lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    lora_dropout=0.1,
    target_modules=["query", "value", "key"],  # for QA models
    task_type=TaskType.QUESTION_ANS,
)
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()
```

```python
# ============================================================


# ============================================================
# 3ï¸âƒ£ Load SQuAD Dataset
# ============================================================
dataset = load_dataset("squad_v2")

# Use small subset for quick demo/testing
train_dataset = dataset["train"]
eval_dataset = dataset["validation"].select(range(1000))
```

```python
len(train_dataset),len(eval_dataset)
```

```python
# ============================================================
# 4ï¸âƒ£ Preprocess Function
# ============================================================
max_length = 384
doc_stride = 128

def preprocess_function(examples):
    questions = [q.strip() for q in examples["question"]]
    inputs = tokenizer(
        questions,
        examples["context"],
        max_length=max_length,
        truncation="only_second",
        stride=doc_stride,
        return_overflowing_tokens=True,
        return_offsets_mapping=True,
        padding="max_length",
    )

    sample_map = inputs.pop("overflow_to_sample_mapping")
    offset_mapping = inputs.pop("offset_mapping")

    start_positions = []
    end_positions = []

    for i, offsets in enumerate(offset_mapping):
        input_ids = inputs["input_ids"][i]
        cls_index = input_ids.index(tokenizer.cls_token_id)
        sequence_ids = inputs.sequence_ids(i)
        sample_idx = sample_map[i]
        answers = examples["answers"][sample_idx]

        if len(answers["answer_start"]) == 0:
            start_positions.append(cls_index)
            end_positions.append(cls_index)
        else:
            start_char = answers["answer_start"][0]
            end_char = start_char + len(answers["text"][0])
            token_start_index = 0
            while sequence_ids[token_start_index] != 1:
                token_start_index += 1
            token_end_index = len(input_ids) - 1
            while sequence_ids[token_end_index] != 1:
                token_end_index -= 1
            if not (
                offsets[token_start_index][0] <= start_char
                and offsets[token_end_index][1] >= end_char
            ):
                start_positions.append(cls_index)
                end_positions.append(cls_index)
            else:
                while token_start_index < len(offsets) and offsets[token_start_index][0] <= start_char:
                    token_start_index += 1
                start_positions.append(token_start_index - 1)
                while offsets[token_end_index][1] >= end_char:
                    token_end_index -= 1
                end_positions.append(token_end_index + 1)

    inputs["start_positions"] = start_positions
    inputs["end_positions"] = end_positions
    return inputs

tokenized_train = train_dataset.map(preprocess_function, batched=True, remove_columns=dataset["train"].column_names)
tokenized_eval = eval_dataset.map(preprocess_function, batched=True, remove_columns=dataset["validation"].column_names)
```

```python
# ============================================================
# 5ï¸âƒ£ Prepare Val Dataloader for BI Computation
# ============================================================
val_loader = DataLoader(
    tokenized_eval,
    batch_size=4,
    collate_fn=default_data_collator,
)

# ============================================================
# 6ï¸âƒ£ Initialize Adaptive LoRA Callback
# ============================================================
adaptive_callback = AdaptiveLoRACallback(
    total_rank=128,
    tau=0.8,
    val_dataloader=val_loader,
    verbose=True
)
```

```python
# ============================================================
# 7ï¸âƒ£ Offline-Safe Evaluation Metrics (Exact Match + F1)
# ============================================================

def compute_metrics(eval_pred):
    """
    Computes Exact Match (EM) and F1 score manually without evaluate.load().
    Works offline and reproduces the same SQuAD metric behavior approximately.
    """
    predictions, labels = eval_pred
    start_logits, end_logits = predictions
    start_positions, end_positions = labels

    # Convert logits â†’ token indices
    start_pred = np.argmax(start_logits, axis=1)
    end_pred = np.argmax(end_logits, axis=1)

    # Exact match (when predicted span matches gold span)
    em = np.mean((start_pred == start_positions) & (end_pred == end_positions)) * 100

    # F1 score based on token overlap between predicted and gold spans
    f1_scores = []
    for s_pred, e_pred, s_gold, e_gold in zip(start_pred, end_pred, start_positions, end_positions):
        # Compute intersection and union of predicted vs gold spans
        pred_tokens = set(range(s_pred, e_pred + 1))
        gold_tokens = set(range(s_gold, e_gold + 1))
        common = len(pred_tokens & gold_tokens)

        if len(pred_tokens) == 0 or len(gold_tokens) == 0:
            f1 = 0
        else:
            precision = common / len(pred_tokens)
            recall = common / len(gold_tokens)
            f1 = (2 * precision * recall) / (precision + recall + 1e-8)
        f1_scores.append(f1)

    f1 = np.mean(f1_scores) * 100
    return {"exact_match": em, "f1": f1}
```

```python
# ============================================================
# 8ï¸âƒ£ Training Setup
# ============================================================
training_args = TrainingArguments(
    output_dir="./qa_results",
    num_train_epochs=3,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    evaluation_strategy="epoch",
    logging_steps=10,
    report_to="none",
    save_strategy="no",
    learning_rate=2e-5,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_eval,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
    callbacks=[adaptive_callback],
    data_collator=default_data_collator,
)
```

```python
print("\nğŸš€ Starting Adaptive LoRA QA Training...")
trainer.train()
```

```python
print("\nğŸ” Final Evaluation on Validation Split...")
metrics = trainer.evaluate()
print(metrics)
```

```python


# ============================================================
# ğŸ§ª 11ï¸âƒ£ Example Prediction
# ============================================================
context = "The Eiffel Tower is located in Paris and was completed in 1889."
question = "Where is the Eiffel Tower located?"

inputs = tokenizer(question, context, return_tensors="pt").to(model.device)
with torch.no_grad():
    outputs = model(**inputs)
start_idx = torch.argmax(outputs.start_logits)
end_idx = torch.argmax(outputs.end_logits)
answer = tokenizer.decode(inputs["input_ids"][0][start_idx : end_idx + 1])

print(f"\nâœ¨ Example Prediction: {answer}")
```
